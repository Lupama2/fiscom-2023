# Autómatas Celulares

Los autómatas celulares consisten en una red de celdas (típicamente una cuadrícula), cada una de las cuales puede adoptar un número discreto y *finito* de estados. El estado de cada celda en un momento dado está determinado por su propio estado y los estados de sus celdas vecinas en el paso de tiempo anterior, según una regla específica.

Matemáticamente, un autómata celular se puede definir a partir de tres componentes $(G$, $E$ y $R)$, donde:
-  $G$: Red o cuadrícula, generalmente unidimensional, bidimensional o tridimensional.
-  $E$: Conjunto *finito* de posibles estados para cada celda.
-  $R$: Regla de transición que define cómo cambia el estado de una celda en función de su propio estado y los estados de otras celdas en la red (típicamente las celdas vecinas).

La dinámica de un autómata celular está determinada completamente por la regla de transición $ R $. El sistema evoluciona en pasos de tiempo discretos, y el estado de cada celda en el tiempo $t+1$ está dado por:

$$ u_{i}(t+1) = R(u_{i}(t), u_{i-1}(t), u_{i+1}(t), \ldots) $$

donde  $u_{i}(t)$ es el estado de la celda $ i$ en el tiempo $t $, y las celdas vecinas se definen de acuerdo con la topología específica de la red.

Los autómatas celulares proporcionan un marco poderoso y flexible para modelar sistemas complejos en la física. Al discretizar el espacio, el tiempo y los valores que toma la variable independiente, ofrecen una forma computacionalmente eficiente de explorar fenómenos de dinámica no lineal y emergentes. Como versión discreta de las Ecuaciones en Derivadas Parciales, cierran la brecha entre los modelos matemáticos continuos y la naturaleza inherentemente discreta de las simulaciones.

#### Aplicaciones en Física

Los autómatas celulares se pueden aplicar a la la simulación de fluidos, el tratamiento del modelo de Ising y otros modelos de magnetismo, el modelado del crecimiento de granos, la descripción de transiciones de fase y otros comportamientos de materiales.






Los llamados autómatas celulares elementales se componen de una línea de celdas con dos estados posibles: 0 y 1. En un autómata celular elemental, el estado de una celda en el siguiente paso de tiempo (iteración) depende únicamente de su propio estado y el estado de sus dos vecinos inmediatos, el de la izquierda y el de la derecha.

Se define un conjunto de reglas que determinan cómo cambian los estados de las celdas en función de sus vecinos. En un autómata celular elemental, hay 8 posibles configuraciones de tres celdas contiguas (cada una de las tres celdas puede estar en estado 0 o 1). Dado que hay dos estados posibles para la celda central en el siguiente paso (0 o 1), hay $2^8$ = 256 reglas posibles. Estas reglas se pueden identificar por números enteros del 0 al 255, que al ser escritos en base 2 dan la representación de la regla.

Por ejemplo la Regla 30 (en binario 00011110 porque $30 = 2^0*0 + 2^1*1 + 2^2*1+ 2^3*1 + 2^4*1 +2^5*0 +2^6*0+ 2^7 * 0$) se preseta en la siguiente tabla:
\begin{array}{|c|c|c||c|}
\hline
\text{Vecino izquierdo} & \text{Celda central} & \text{Vecino derecho} & \text{Nuevo estado, Regla 30} \\
\hline
1 & 1 & 1 & 0 \\
\hline
1 & 1 & 0 & 0 \\
\hline
1 & 0 & 1 & 0 \\
\hline
1 & 0 & 0 & 1 \\
\hline
0 & 1 & 1 & 1 \\
\hline
0 & 1 & 0 & 1 \\
\hline
0 & 0 & 1 & 1 \\
\hline
0 & 0 & 0 & 0 \\
\hline
\end{array}


Algunos ejemplos muy conocidos de autómatas celulares elementales son la Regla 30, la Regla 110 y la Regla 90. Cada uno de estos autómatas produce patrones únicos a partir de una condición inicial dada. Los autómatas celulares elementales pueden exhibir comportamientos que van desde simples y repetitivos hasta complejos y aparentemente aleatorios. Algunos autómatas, como la Regla 110, se pueden usar, eligiendo adecuadamente la condición inicial, como una computadora universal. El matemático británico Matthew Cook fue el primero en demostrar esto en 1998, y su demostración fue revisada y simplificada por Stephen Wolfram en su libro "A New Kind of Science" (2002).

Para más información ver [Wikipedia](https://en.wikipedia.org/wiki/Elementary_cellular_automaton#) o el libro de Wolfram, "A new kind of Science"...


Los autómatas celulares elementales son un modelo matemático simple pero poderoso para estudiar la dinámica de sistemas complejos y la emergencia de patrones a partir de reglas locales simples. Han sido aplicados en diversas áreas de investigación, como la física, la biología y la ciencia de la computación.
<h3> Clasificación de Wolfram </h3>

La clasificación de Wolfram se basa en la observación del comportamiento de los autómatas celulares a partir de condiciones iniciales simples, como una sola celda en estado "1" rodeada de celdas en estado "0". Wolfram identificó cuatro clases principales de comportamiento:

1. Clase 1: Los autómatas celulares de esta clase evolucionan rápidamente hacia un estado homogéneo y uniforme. Casi todos los patrones iniciales convergen en un estado fijo en el que todas las celdas tienen el mismo valor. Un ejemplo de un autómata celular de Clase 1 es la Regla 4 (00000100).

2. Clase 2: Los autómatas celulares de Clase 2 producen patrones estables, periódicos o repetitivos. Aunque pueden presentar estructuras más complejas que los de Clase 1, estos patrones siguen siendo simples y predecibles. La Regla 90 (01011010) es un ejemplo de autómata celular de Clase 2.

3. Clase 3: Los autómatas celulares de Clase 3 generan patrones aparentemente aleatorios y caóticos. A partir de condiciones iniciales simples, estos autómatas pueden producir una amplia variedad de comportamientos complejos que pueden parecer estocásticos, aunque sean deterministas. La Regla 30 (00011110) es un ejemplo de autómata celular de Clase 3.

4. Clase 4: Los autómatas celulares de Clase 4 son los más interesantes y complejos. Generan patrones que involucran una mezcla de regularidad y aleatoriedad, y pueden presentar estructuras localizadas que se mueven y/o interactúan entre sí, como deslizadores o "gliders". La Regla 110 (01101110) es un ejemplo de autómata celular de Clase 4.


Para interpretar los resultados de los autómatas celulares en las gráficas generadas con el código más abajo:

1. Eje vertical (Paso): El eje vertical representa el tiempo o la cantidad de iteraciones. Cada fila en la gráfica muestra el estado de las celdas en un paso específico.

2. Eje horizontal (Índice de celda): El eje horizontal representa las celdas individuales en el autómata celular. Cada columna representa una celda específica.

3. Las celdas en blanco representan el estado "0" y las celdas en negro representan el estado "1". Los patrones que se observan en la gráfica muestran cómo cambian los estados de las celdas a lo largo del tiempo según las reglas del autómata celular.


- Regla 30: Genera patrones aparentemente aleatorios y caóticos a partir de una condición inicial simple (un solo "1" rodeado de "0"). A pesar de ser un sistema determinista, la Regla 30 puede generar comportamientos complejos y se puede usar como generador de números pseudoaleatorios.
- Regla 90: genera un fractal conocido como triángulo de Sierpiski.
- Regla 110: Muestra comportamientos más estructurados en comparación con la Regla 30, pero también tiene áreas de aparente aleatoriedad. La Regla 110 es especialmente interesante porque es capaz de realizar cómputo universal, lo que significa que, en teoría, puede simular cualquier máquina de Turing. Esto implica que la Regla 110 es un ejemplo de un sistema simple que puede realizar cálculos complejos y modelar sistemas complejos.

En resumen, los resultados de los autómatas celulares se interpretan como patrones generados a lo largo del tiempo según las reglas y condiciones iniciales especificadas. Estos patrones pueden revelar características interesantes y ayudarnos a comprender mejor los sistemas complejos, la computación y otros fenómenos.





```python
import numpy as np
import matplotlib.pyplot as plt

def regla(numero, celda_izquierda, celda_central, celda_derecha):
    # número de regla como un número binario de 8 bits
    regla_binaria = f"{numero:08b}"
    mapa_regla = {
        "111": regla_binaria[0],
        "110": regla_binaria[1],
        "101": regla_binaria[2],
        "100": regla_binaria[3],
        "011": regla_binaria[4],
        "010": regla_binaria[5],
        "001": regla_binaria[6],
        "000": regla_binaria[7],
    }
    #aplica la regla dada la configuración de las 3 celdas
    return mapa_regla[f"{celda_izquierda}{celda_central}{celda_derecha}"]

def ejecutar_ac(numero_regla, estado_inicial, pasos):
    estado = estado_inicial
    historial = [estado]
    for _ in range(pasos):
        siguiente_estado = []
        #recorre todas las celdas y sus vecinas usando condiciones periódicas de contorno
        for i in range(len(estado)):
            izquierda = estado[(i - 1) % len(estado)]
            centro = estado[i]
            derecha = estado[(i + 1) % len(estado)]
            #resultado de la regla en la celda correspondiente del nuevo estado.
            siguiente_estado.append(regla(numero_regla, izquierda, centro, derecha))
        estado = siguiente_estado
        historial.append(estado)
    return historial

def graficar_ac(numero_regla, estado_inicial, pasos):
    historial = ejecutar_ac(numero_regla, estado_inicial, pasos)
    #genera una matriz con todos los estados
    datos = np.array(historial, dtype=int)
    plt.imshow(datos, cmap='binary')
    plt.title(f"Regla {numero_regla}")
    plt.xlabel("Índice de celda")
    plt.ylabel("Paso")
    plt.show()

# Regla 4
numero_regla = 4
#lista con el estado de cada celda
estado_inicial = ["0"] * 100 + ["1"] + ["0"] * 100
pasos = 100
graficar_ac(numero_regla, estado_inicial, pasos)

# Regla 30
numero_regla = 30
#lista con el estado de cada celda
estado_inicial = ["0"] * 100 + ["1"] + ["0"] * 100
pasos = 100
graficar_ac(numero_regla, estado_inicial, pasos)

# Regla 90
numero_regla = 90
estado_inicial = ["0"] * 100 + ["1"] + ["0"] * 100
pasos = 100
graficar_ac(numero_regla, estado_inicial, pasos)

# Regla 110
numero_regla = 110
estado_inicial = ["0"] * 100 + ["1"] + ["0"] * 100
pasos = 100
graficar_ac(numero_regla, estado_inicial, pasos)


```
