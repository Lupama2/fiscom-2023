```python
#@title Serie de Taylor
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, lambdify, sin, diff, series, latex

x = symbols('x')

# Definimos la función f su desarrollo en serie hasta orden n en el punto x0
n = 8
x0 = 0

# Para poner otra función hay que importarla de sympy (ver más arriba)
f = sin(sin(x))


taylor = series(f, x, x0, n ).removeO().expand().evalf()
# removeO elimina la indicación del Orden de la serie
# evalf evalua todo lo que puede evaluar
# expand hace la expansión del polinomio

#taylor = series(f, x, x0, n ).removeO()

# Etiquetas para el gráfico
labelf = "$"+latex(f) +"$"
labelt = "$"+latex(taylor) +"$"

# Esto permite evaluar las funciones "f" y "taylor"
f1 = lambdify(x, f)
tay = lambdify(x, taylor, "numpy")

# Rango de valores de x a evaluar
x = np.arange(0,6,0.01)

plt.plot(x, f1(x),  color='red', label=labelf)
plt.plot(x, tay(x),  color='blue', label=labelt)

# Límites del gráfico
plt.xlim(0,6)
plt.ylim(-1,1)

plt.legend(loc ="upper left")
plt.grid()
plt.show()


```
