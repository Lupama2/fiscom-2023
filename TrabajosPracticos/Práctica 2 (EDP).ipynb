## Ejercicio 1

El modelo de FitzHugh-Nagumo representa la propagación del impulso nervioso. Dependiendo de los parámetros, también muestra la formación de patrones periódicos espaciales estacionarios, llamados *estructuras de Turing* (quien propuso que modelos de este tipo podían explicar la formación de estructuras en el desarrollo embrionario).

Considere la siguiente forma del modelo:
$$
\frac{\partial u}{\partial t} = D_u \frac{\partial^2 u}{\partial x^2} + g_1(u,v),~~~~~~~~~~~~~(1)
\\
\frac{\partial v}{\partial t} = D_v \frac{\partial^2 v}{\partial x^2} + g_2(u,v),~~~~~~~~~~~~~~(2)
$$
con las siguientes interacciones entre ambas vbariables:
$$
g_1(u,v) = u-v^3-v+a,~~~~~~~~~~~~~(3)
\\
g_2(u,v) = b(u-v),,~~~~~~~~~~~~~~~~~~~~~~(4)
$$
donde $u = u(x,t)$ y $v = v(x,t)$.

Programe la solución del sistema en un dominio periódico, discretizado con una grilla uniforme en el espacio y el tiempo. Use un método FTCS, asegurando la continuidad del flujo difusivo en el borde del dominio. Use los siguientes parámetros: $D_u = 1$, $D_v = 100$, $a = 0.005$, $b = 10$. Para la discrtización, considere indicativamente los siguientes parámetros: $L = 100$, $\Delta x = 1$, $\Delta t = 10^{-3}$, o use su mejor criterio.

1. Compruebe que una perturbación inicial homogénea del equilibrio homogéneo es estable.
2. Usando una perturbación aleatoria de pequeña amplitud, obtenga el patrón estacionario.
3. Reduciendo el valor del parámetro $b$ (\~1), observe que la longitud de onda del patrón de Turing aumenta. Agrandándolo (\~50), observe que el patrón regular no se forma (o se forma muy lentamente, por ejemplo para $b \approx 40$).
4. Extienda la solución a un recinto cuadrado (para los parámetros que considere mejores), y calcule los patrones estacionarios 2D que resultan.



##Ejercicio 2

La *ecuación de Fisher-Kolmogoroff* combina una dinámica logística (tema que veremos próximamente en más detalle) con un transporte difusivo, en forma de una ecuación de reacción-difusión. En 1D tenemos:
$$
\frac{\partial u}{\partial t} = r u (1-u) + D \frac{\partial^2 u}{\partial^2 x},
$$
donde  $u = u(x,t)$.

Esta ecuación admite soluciones en forma de ondas viajeras, lo cual es un poco sorprendente, siendo que no es una ecuación de ondas: la derivada temporal del miembro izquierdo es una deivada primera, no segunda. Además, el rol intuitivo de la difusión es el de borronear las inhomogeneidades, no propagarlas! Sin embargo, la acción conjunta del término logístico y el difusivo produce una especie de milagro matemático, y se propagauna onda a velocidad $c = 2 \sqrt{r D}$, sin cambiar de forma.

Analice la evolución de una condición inicial en forma de escalón suave, tipo:
$$
u (x,0) = \frac{1}{2} \biggl( 1 - erf \Bigl( \frac{x-x_0}{L} \Bigr) \biggr),
$$
usando al menos dos valores del parámetro de escala espacial $L$.

La onda viajará hacia la derecha, así que elija sabiamente el valor de x_0, el espacio de integración, y los parámetros $r$ y $D$, para poder verla. Asimismo, defina condiciones de borde compatibles con la condición inicial.

1. Use un método FTCS. Dibuje el perfil de la onda viajera a distintos tiempos. ¿Puede usar cualquier valor de $\Delta x$ y $\Delta t$? Calcule la velocidad del frente de onda (definiéndolo como le parezca mejor) en función del tiempo.

2. Use el método Leapfrog para calcular lo mismo, y señale las diferencias.

3. ¿Cómo denería ser la condición inicial para que la onda viaje hacia la izquierda?
