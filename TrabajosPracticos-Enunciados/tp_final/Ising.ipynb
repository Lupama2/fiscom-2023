- En las simulaciones que acaba de hacer, los nuevos estados son generados modificando los estados previos y por tanto las configuraciones consecutivas están correlacionadas. Es muy útil estimar el tiempo de autocorrelación, que describe la cantidad de pasos de MC necesarios para que dos mediciones no estén correlacionadas. Para esto definimos las  funciones de correlación temporal:
$$C_A(t)= \frac{\langle A(t+t_o)A(t_o)\rangle-\langle A\rangle^2}{\langle A^2\rangle-\langle A\rangle^2},$$
 donde $A$ es una magnitud que depende del tiempo. Los promedios son sobre todos los posibles origenes de tiempo $t_o$. Debido a que el origen del tiempo es arbitrario para un sistema en equilibrio, $C_A$ sólo depende de la diferencia de tiempo $t$ y no de $t$ o $t_o$ por separado. Para tiempos suficientemente largos, $A(t)$ y $A(t_o)$ se descorrelacionan $\langle A(t+t_o)A(t_o)\rangle \rightarrow \langle A(t+t_o)\rangle \langle A(t_o)\rangle=\langle A\rangle^2$  y por tanto  $C_A \rightarrow 0$ cuando $t \rightarrow \infty$. $C_A(t=0)$ está normalizado a 1. En general, $C_A(t)$ decae exponencialmente con $t$, lo que permite definir un *tiempo de correlación*, $\tau_A$, que depende de la cantidad física $A$ y de los par\'ametros del sistema, por ejemplo la temperatura.

 Para $L=10$, calcule las funciones de autocorrelación $C_m$ y $C_e$ para distintas temperaturas tanto por debajo como por encima de $T_c$. Para $T=2.5$, calcule las funciones de autocorrelación $C_m$ para distintos tamaños, $L=5, 10,50$ y $100$. Estime $\tau_A$ en función de temperatura ajustando curvas de la forma $C_A \sim e^{-t/\tau_A}$. ¿Son comparables entre sí los tiempos de correlación $\tau_m$ y $\tau_e$?
