## Agregación limitada por difusión

Simule un proceso de *agregación limitada por difusión*. Esto es, se tiene un conjunto partículas que difunden en un medio y la interacción entre ellas puede hacer que queden pegadas entre sí formando un agregado.

Para simplificar el modelo considere a las partículas como caminantes aleatorios en una red cuadrada con condiciones de borde periódicas. Además, tome como condición inicial una partícula fija en el centro de la red. Las otras partículas se pueden pegar con esa o con otras otras que ya hayan quedado pegadas. Una partícula queda pegada (se fija su posición) si hace contacto (llega a una posición vecina) con otra partícula ya fija a lo largo de los enlaces de la red o a 45$^{\circ}$.

Indicativamente, use una red de
1024 × 1024 y del orden de 2 × $10^4$
caminantes aleatorios que parten de posiciones distribuidas uniformemente.

- Haga un gráfico de la estructura que se forma, coloreando las partículas de acuerdo al orden en que se pegaron.

- Usando círculos de radio $r$ con centro en la semilla, calcule el número de partículas $N(r)$ contenidas en el círculo en función del radio. Identifique en un gráfico $\log-\log$
de $N (r)$ los tres regímenes que aparecen en función de $r$.
En el régimen de valores de $r$ intermedios se debería tener
$$
N(r)\propto r^D
$$

- Calcule la dimensión fractal $D$ de la estructura dendrítica resultante. Observe que, para obtener un resultado sensato, los radios tienen que abarcar al menos un orden de magnitud en la región de valores intermedios de $r$.

- Implemente el método de *box-counting* para calcular la dimensión fractal. Para ello, divida la red en "cajas" del mismo tamaño $\ell$ y cuente cuántas cajas $B(\ell)$ son necesarias para cubrir el agregado (cuántas de las cajas tienen partículas en su interior).

- Grafique la cantidad de cajas, $B(\ell)$, necesaria para cubrir la estructura, en función
de $1/\ell$; una
posible elección de $\ell$ son potencias de $2$, usando cajas el doble de grandes
en cada escala). En el gráfico $\log-\log$ de $B (\ell)$ vs $1/\ell$ identifique nuevamente
tres regímenes y calcule la dimensión fractal $D$ en la región intermedia de valores de $1/\ell$ donde el comportamiento es del tipo ley de potencias ($\propto(1/\ell)^D$). Vale la
misma observación que con el otro método acerca del rango de magnitudes que tiene que abarcar $1/\ell$
para obtener un buen resultado.

- Calcule la dimensión *box-counting* para un ensemble de agregados del mismo tamaño.
