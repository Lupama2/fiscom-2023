###Indicaciones para trabajo final

La entrega del trabajo final debe consistir de:

- Un archivo pdf con el informe.

- Material suplementario.

- El informe debe estar redactado como un artículo, de manera formal, conteniendo:

 - Título

 - Autor

 - Fecha

 - Resumen

 - Introducción: Planteo del problema y de la metodología a utilizar.

 - Resultados: Presentación de los resultados, de manera lógica y entendible, ilustrados con gráficos y explicaciones de las condiciones (parámetros, etc) que se usaron para obtenerlos. Los gráficos deben estar numerados, tener un caption o leyenda que explique lo que muestra y brevemente los parámetros. Los textos en la figura deben ser legibles (no microscópicos), y si hay leyendas (colores, símbolos, etc) tiene que entenderse qué representan. Todos los gráficos tienen que estar comentados en el texto. Pueden hacerse observaciones y comentarios sobre los resultados, no es necesario dejar toda la discusión para el final.

 - Discusión y conclusiones: Recapitulación de los resultados que se obtuvieron y conclusiones que puedan obtener. Discusión sobre cosas que podrían haber hecho distinto.

 - Bibliografía

Por favor, el informe debe estar correctamente redactado y sin errores de ortografía ni de tipeo. Revísenlo antes de entregar.

El material suplementario contendrá el código utilizado, incluyendo bibliotecas y archivos de inicialización. También pueden agregar videos u otro material que no pudieran, o no quisieran, poner en el informe. En tal caso, en el informe comentarán en el lugar adecuado que en el Material Suplementario se puede ver tal y cual cosa.

### Indicaciones para la presentación final

- La presentación es de un máximo **estricto** de 15 minutos  y debe consistir de
    - una breve introducción al problema.
    - una breve descripción de los métodos utilizados y de la estrategia computacional elegida.
    - una presentacion de los principales resultados obtenidos.
    - conclusiones, incluyendo posibles continuaciones del trabajo realizado.

- El tiempo reducido seguramente va a obligar a hacer una selección de los resultados a presentar. A veces **más** termina siendo **menos**: en general es mejor presentar pocas cosas de manera clara que muchas a las apuradas.
- En una charla de 15 minutos es muy difícil explicar con mucho detalle.
- Las preguntas posteriores a la presentación se van a centrar en el trabajo realizado pero **pueden abarcar diferentes temas de la materia e incluir preguntas sobre el código utilizado**.
