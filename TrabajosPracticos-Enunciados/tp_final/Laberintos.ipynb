#Laberintos

Una posible manera de resolver un laberinto sería soltar algunos miles de partículas en la entrada, y esperar hasta que alguna de ellas cruce la salida. Una vez identificada la primera que lo haga, su trayectoria mostrará el recorrido que resuelve el laberinto. Habría distintas posibilidades:

- Un ensemble de partículas difusivas no interactuantes, haciendo cada una un random walk.
- Un ensemble de partículas balísticas, chocando contra las paredes.
- Un ensemble de partículas con interacciones tipo Lennard-Jones, como si fuera una nube de gas que va llenando el laberinto.

Discuta los pros y contras de cada una de las opciones, elija la que más le guste, impleméntela y analice los resultados. Explore distintos tamaños del ensemble, velocidades (o sea, temperatura), interacciones, etc., tratando de optimizar la resolución.

Observe que en los dos casos conservativos, en principio, no necesitaría guardar las trayectorias de todas las partículas, podría invertir las velocidades cuando la ganadora llegue a la salida y seguir su rastro. ¿Funcionará?
