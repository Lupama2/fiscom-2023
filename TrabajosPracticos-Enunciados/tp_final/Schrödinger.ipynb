##Ecuación de Schrödinger


Estudie la dinámica cuántica de una partícula en un potencial $V(x)$, integrando numéricamente la ecuación de Schrödinger:
$$
i\hbar\frac{\partial\psi}{\partial t} = \left[-\frac{\hbar^2}{2m}\frac{\partial^2}{\partial
x^2} + V(x)\right]\psi,
$$
Como condición inicial, use un paquete
de onda gaussiano con vector de onda promedio $k_0$, centrado en $x_0$ y de ancho $\sigma$,
$$
\psi(x,t=0)= C e^{ik_0x} e^{-(x-x_0)^2/4\sigma^2}
$$
y use unidades tales que $m=1$ y $\hbar=1$.

Calcule numéricamente la evolución del paquete de onda $\psi$ en el recinto $[0,L]$ en los siguientes casos.

- Partícula libre: $V(x)=0$; $k_0=50\pi$.
- Barrera de potencial:
	$V(x)=
	\begin{cases}
		V_0 \text{ si } |x-L/2|<a, \\
		0  \text{  si } |x-L/2|>a,
	\end{cases}$ \\
	con $a=0.032$ y $V_0=(50\pi)^2$.
	1. $k_0^2/2 < V_0$ (e.g. $k_0=50\pi$),
	2. $k_0^2/2\lesssim V_0$ (e.g. $k_0=70\pi$),
	3. $k_0^2/2 > V_0$ (e.g. $k_0=100\pi$).
- Barrera doble de potencial en función de la distancia entre las barreras.
- Potencial periódico en función de la longitud de onda del potencial.
- Potencial desordenado. Elegir el potencial en cada punto del espacio aleatoriamente de una distribución uniforme $[0,W]$ para diferentes valores de $W$.

Elija como condiciones de contorno $\psi(0,t)=\psi(L,t)=0$, con $L=1$, $x_0=1/4$, $\sigma_0=0.05$, $\hbar=1$. Para la discretización, use por ejemplo: $\Delta x = 0.001$ y $\Delta t= 2\Delta x^2$.

En cada caso obtenga primero la constante de normalización $C$. Luego, durante un tiempo $t\lesssim (L/2)/(k_0)$, grafique en función del tiempo:

- La energía: $E(t)=\langle H\rangle$.
- La norma de la función de onda: $\int |\psi(x,t)|^2\,dx$.
- La posición media $\langle x \rangle$ y su dispersión $(\delta x)^2=\langle x^2\rangle -\langle x \rangle^2$.
- El momento medio $\langle p \rangle$ y su dispersión $(\delta p)^2=\langle p^2 \rangle -\langle p \rangle^2$.

Verifique que se conservan la energía y la norma. Luego grafique $|\psi(x,t)|^2$ vs. $x$ para varios instantes.

Recuerde: $\langle A \rangle \equiv \langle \psi | A | \psi\rangle \equiv \int  \psi^*(x,t)A\psi(x,t)\,dx$.
