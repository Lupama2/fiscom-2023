##Formación de dunas


En 1993, [Ouchi y Nishimori](https://drive.google.com/file/d/1q_OqeWeXiZUpfPiGmlWF-T6-0YdRfh9Y/view?usp=drive_link) estudiaron un par de modelos simples para simular la formación de patrones ondulados y dunas de arena debido al viento. Se tiene un autómata celular bidimensional de dimensiones $L_x\times L_y$ donde cada celda $h(i, j)$ (con $0\leq i \leq L_x$, $0\leq i \leq L_y$) tiene un valor real que corresponde a la diferencia entre la altura de la arena en ese punto y la altura promedio.

### Modelo I

La dinámica está descrita por dos procesos:

1. El *salto* en la cual el viento sopla en el sentido positivo del eje $x$, levanta granos de arena y los mueve hasta otro punto del plano. Para el autómata esto significa que desde cada posición $(x, y)$, una cantidad aleatoria de arena $q \in (0, q_\text{max})$ se deposita en la posición $(x + L, y)$, conservando la cantidad total de arena:
	\begin{equation}
		\begin{aligned}
			h(x, y) &\longrightarrow h(x, y) - q,\\
			h(x + L, y) &\longrightarrow h(x + L, y) + q.
		\end{aligned}
	\end{equation}
	La longitud del  *salto* $L$ depende de la fuerza con que sopla el viento y de la altura de la posición inicial. Se supone una dependencia lineal:
	\begin{equation}
	 	L = L_0 + b~h(x, y),
	 \end{equation}
	donde la dependencia con la posición inicial toma en cuenta el tiempo de vuelo de cada partícula.
1. El segundo proceso es la *relajación*, por la cual los granos de arena tienden a caer por las pendientes. Si la diferencia entre alturas vecinas $h(x, y)$ y $h(x', y')$ sobrepasa un valor umbral $\Delta$, entonces una cantidad aleatoria de arena, $r$, cae de un lugar al otro simulando la acción de la gravedad:
	\begin{equation}
		\begin{aligned}
			h(x, y) &\longrightarrow h(x, y) - r,\\
			h(x', y') &\longrightarrow h(x', y') + r.
		\end{aligned}
	\end{equation}
	La cantidad $r$ tiene que ser menor o igual que el gradiente que la generó para asegurar que no se invierta el gradiente en el proceso.


Este modelo sirve para simular la formación de ondas en la superficie de la arena; para la formación de dunas se utiliza otro.

###Modelo II

El segundo modelo sólo tiene dos variaciones del primero.


1. En barlovento (el lado de la duna en que choca el viento) los granos levantados chocarán rápidamente con la duna y $L$ será corto; lo contrario pasará por el lado de  sotavento. Entonces se redefine $L$ como:
	\begin{equation}
		L = L_0 - b ~ \tanh(\nabla \cdot h(x, y)).
	\end{equation}
2. La probabilidad de que otro grano sea levantado por el choque de un grano inicial es mayor en el lado de la duna que da al viento; entonces redefinimos:
	\begin{equation}
		q = q_0 + b' \cdot \tanh(\nabla \cdot h(x, y)),
	\end{equation}
	donde $q_0$ y $b'$ son constantes.
